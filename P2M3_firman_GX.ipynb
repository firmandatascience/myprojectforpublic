```python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'firman-p2m3-'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'firman-p2m3-'
path_to_data = '/content/P2M3_firman_data_cleaned.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-trip-dataset'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

Data to be processed.

```python
# Expectation 1 : value in id is unique

validator.expect_column_values_to_be_unique(column='id')
```

Expectation 1 is completed.

```python
# Expectation 2 : Rating between 1 and 5

validator.expect_column_values_to_be_between(
    column='rating', min_value=1, max_value=5
)
```

Expectation 2 is completed.

```python
# Expectation 3 : recommended_ind to be in set 0-1

validator.expect_column_values_to_be_in_set(
    column='recommended_ind', value_set=[0,1]
)
```

Expectation 3 is completed.

```python
#Expectation 4 : Division should be in list ('General', 'General Petite', 'Initmates')

validator.expect_column_values_to_be_in_set(
    column='division_name', value_set=['General', 'General Petite', 'Initmates']
)
```

Expectation 4 is completed.

```python
# Expectation 5 : column recommended_ind must be string

validator.expect_column_values_to_be_of_type(column='department_name', type_='str')

```

Expectation 5 is completed.

```python
# Expectation 6 : column age is integer

validator.expect_column_values_to_be_of_type(column='age', type_='int64')
```

Expectation 6 is completed.

```python
# Expectation 7 : No missing value in age column

validator.expect_column_values_to_not_be_null(column='age')
```

Expectation 7 is completed.
